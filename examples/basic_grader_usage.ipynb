# RFTKit - Basic Grader Usage

This notebook demonstrates how to use RFTKit's grader system for OpenAI's Reinforcement Fine-Tuning.

We'll cover:
1. Creating individual graders (Formatting, Rubric)
2. Exporting grader configs for OpenAI's RFT API

## Setup


```python
from rftkit import (
    FormattingGrader,
    RubricGrader,
    RubricItem,
    export_grader_config,
)
import json
```

## 1. Creating a FormattingGrader

The FormattingGrader validates the JSON structure of model outputs.


```python
# Create a formatting grader
formatting_grader = FormattingGrader(
    name="json_validator",
    weight=0.2  # 20% weight in final score
)

# View the config
print(json.dumps(formatting_grader.config, indent=2))
```

## 2. Creating a RubricGrader

The RubricGrader uses an LLM to evaluate outputs against specific criteria.

```python
# Define a rubric
rubric = RubricItem(
    id=1,
    content="Assessment Validity: How accurately the output addresses the prompt. 0.0 = completely wrong, 1.0 = perfect"
)

# Create a rubric grader
rubric_grader = RubricGrader(name="rubric_validity", rubric=rubric, weight=1.0)

# View the config
print(json.dumps(rubric_grader.config, indent=2))
```
