# RFTKit - Basic Grader Usage

This notebook demonstrates how to use RFTKit's grader system for OpenAI's Reinforcement Fine-Tuning.

We'll cover:
1. Creating individual graders (Formatting, Indicator, Rubric)
2. Combining graders with AggregatorGrader
3. Exporting grader configs for OpenAI's RFT API


## Setup


```python
from rftkit import (
    FormattingGrader,
    IndicatorGrader,
    RubricGrader,
    AggregatorGrader,
    RubricItem
)
import json
```

## 1. Creating a FormattingGrader

The FormattingGrader validates the JSON structure of model outputs.


```python
# Create a formatting grader
formatting_grader = FormattingGrader(
    name="json_validator",
    weight=0.2  # 20% weight in final score
)

# View the config
print(json.dumps(formatting_grader.config, indent=2))
```

## 2. Creating an AggregatorGrader

The AggregatorGrader combines all graders using the cluster-based formula.


```python
# Define a simple rubric
rubric = RubricItem(
    id=1,
    content="Assessment Validity: How accurately the activity measures the intended skill."
)

# Create graders
rubric_grader = RubricGrader(name="rubric_validity", rubric=rubric, weight=1.0)
indicator = IndicatorGrader(name="indicator_validity", rubric_name="assessment_validity")

# Aggregate
aggregator = AggregatorGrader(
    name="final_score",
    graders=[formatting_grader, rubric_grader, indicator],
    rubric_to_indicator_map={"rubric_validity": "indicator_validity"},
    formatting_grader_name="json_validator"
)

print("Formula:", aggregator.config["calculate_output"])
```
